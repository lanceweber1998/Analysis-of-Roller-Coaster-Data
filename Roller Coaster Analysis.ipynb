## Rank Over Year Function (One Roller Coaster)
The function below take three arguments: a roller coaster name called **'coaster_name'**, an amusement park name called **'park'**, and a dataframe called **'data'**. Input the info to get the ranking of each roller coaster at it's corresponding park over the years!

```python
import matplotlib.pyplot as plt

def rank_over_years(coaster_name, park, data):
    group_by_coaster = data[data.Name == coaster_name]
    group_by_park = group_by_coaster[group_by_coaster.Park == park].reset_index()
    
    # creating X and Y values for line graph
    x = list(group_by_park['Year of Rank'])
    y = list(group_by_park['Rank'])
    
    #setting up graph
    fig, ax = plt.subplots()
    ax.set_yticks(y)
    ax.set_xticks(x)
    plt.plot(x, y)
    plt.xlabel('Year')
    plt.ylabel('Rank')
    plt.title(coaster_name + ' Rank Over Time At ' + park)
    plt.show()
    plt.clf()
```

## Examples:

```python
rank_over_years('Apollo’s Chariot', 'Busch Gardens Williamsburg', steel)
```

```python
rank_over_years('The Voyage', 'Holiday World', wood)
```

```python
rank_over_years('Bizarro', 'Six Flags New England', steel)
```

## Rank Over Years Function (Two Roller Coasters)
This function is like the One Roller Coaster function but take two roller coasters and compares their ranks over the years at a a selected park for each roller coaster

```python
def multi_rank_over_years(coaster_name_01, coaster_name_02, park_01, park_02, data):
    group_by_coaster_01 = data[data.Name == coaster_name_01]
    group_by_park_01 = group_by_coaster_01[group_by_coaster_01.Park == park_01].reset_index()
    group_by_coaster_02 = data[data.Name == coaster_name_02]
    group_by_park_02 = group_by_coaster_02[group_by_coaster_02.Park == park_02].reset_index()
    
    # creating X and Y values for line graph
    x1 = list(group_by_park_01['Year of Rank'])
    x2 = list(group_by_park_02['Year of Rank'])
    y1 = list(group_by_park_01['Rank'])
    y2 = list(group_by_park_02['Rank'])
    
    #setting up graph
    fig, ax = plt.subplots()
    #ax.set_yticks(combined_ranks)
    #ax.set_xticks(x)
    plt.plot(x1, y1, label=coaster_name_01)
    plt.plot(x2, y2, label=coaster_name_02)
    plt.xlabel('Year')
    plt.ylabel('Rank')
    plt.title(f'{coaster_name_01}  Rank At  {park_01} and {coaster_name_02} Rank At {park_02}')
    plt.legend()
    plt.show()
    plt.clf()
```

## Examples

```python
multi_rank_over_years('Boulder Dash', 'El Toro', 'Lake Compounce', 'Six Flags Great Adventure', wood)
```

```python
multi_rank_over_years('Millennium Force', 'Bizarro', 'Cedar Point', 'Six Flags New England', steel)
```

## Top *n'th* place starting at 2013

```python
import matplotlib.pyplot as plt

def top_n(data, n):
    
    # Filter the data to include only the top n rankings
    top_n_rankings = data[data['Rank'] <= n]
    
    # Create a figure and axis for plotting
    fig, ax = plt.subplots(figsize=(10, 10))
    
    # Loop through each unique coaster name in the top n rankings
    for coaster in set(top_n_rankings['Name']):
        # Get the rankings for the current coaster
        coaster_rankings = top_n_rankings[top_n_rankings['Name'] == coaster]
        if not coaster_rankings.empty:
            # Plot the rankings over the years for the current coaster
            ax.plot(coaster_rankings['Year of Rank'], coaster_rankings['Rank'], label=coaster)
    
    # Set the y-axis ticks to be 1 through 5
    ax.set_yticks([i for i in range(1, 6)])
    # Invert the y-axis so that rank 1 is at the top
    ax.invert_yaxis()
    
    # Set the title and labels for the plot
    plt.title(f'Top {n} Rankings')
    plt.xlabel('Year')
    plt.ylabel('Ranking')
    
    # Get handles and labels from the current axes
    handles, labels = ax.get_legend_handles_labels()
    
    # Display the legend outside the plot using the collected handles and labels
    ax.legend(handles=handles, labels=labels, loc='center left', bbox_to_anchor=(1, 0.5))
    
    # Adjust the layout to make room for the legend
    plt.tight_layout(rect=[0, 0, 0.85, 1])
    
    # Show the plot
    plt.show()
    # Close the plot to free up memory
    plt.close()
    
# Example usage (assuming 'data' is a DataFrame with appropriate columns)
# top_n(data, 5)
```

## Examples

#### Top 5 Wooden Roller Coasters

```python
top_n(wood, 5)
```

#### Top 5 Steel Roller Coasters

```python
top_n(steel, 5)
```

### Histograms

Making a function that plots a histogram of any **numeric** column of the roller coaster dataframe

```python
def plot_histogram(data, column):
    plt.hist(data[column].dropna())
    plt.title(f'Histogram of Roller Coaster {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.show()
    plt.clf()
```

### Examples

```python
plot_histogram(rc, 'speed')
```

```python
plot_histogram(rc, 'length')
```

```python
plot_histogram(rc, 'height')
```

Roller Coaster Height is too bunched up, so i will make a new function that fixes this specifically for the **height** column

```python
def plot_height_histogram(df):
    heights = df[df['height'] <= 140]['height'].dropna()
    plt.hist(heights, color='r')
    plt.title('Histogram of Roller Coaster Height Less Than 140ft')
    plt.xlabel('Height')
    plt.ylabel('Count')
    plt.show()
    plt.clf()
```

```python
plot_height_histogram(rc)
```

### Pie Chart

Making a function that creates a pie chart comparing the number of **operating** roller coasters to the number of **closed** roller coasters

```python
def pie_operating(df):
    operating = df[df['status'] == 'status.operating']
    closed = df[df['status'] == 'status.closed.definitely']
    
    num_operating = len(operating)
    num_closed = len(closed)
    proportion = [num_operating, num_closed]
    pie_labels = ['Operating', 'Closed']
    
    plt.pie(proportion, autopct='%0.1f%%', labels=pie_labels)
    plt.axis('equal')
    plt.show()
    plt.clf()
```

```python
pie_operating(rc)
```

### Scatter Plot

Making a scatter plot for any two **numerical** columns in our dataframe

```python
def plot_scatter(df, column_x, column_y):
    plt.scatter(df[column_x],df[column_y])
    plt.title('Scatter Plot of {} vs. {}'.format(column_y,column_x))
    plt.xlabel(column_x)
    plt.ylabel(column_y)
    plt.show()
    plt.clf()
```
